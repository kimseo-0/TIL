`uv pip install scikit-surprise`

파이썬 버전 3.11 이하로 가상환경 재설치

numpy 호환 에러가 있어 버전 2 이상 설치

`uv add "numpy>2"`

`uv add scikit-surprise`

https://visualstudio.microsoft.com/ko/visual-cpp-build-tools/

vs build 툴 설치 후 실행

위 c++ 빌드 도구를 선택 후 설치

🚨 만약 선택을 안하고 했다면?   
vs build 툴을 다시 열고 

수정을 누르면 위와 동일한 창이 나와서 다시 설치를 할 수 있다.
